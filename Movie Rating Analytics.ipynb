```python
#Styling 
sns.set_style('whitegrid')
list1 =[]
mylabel = list()
for gen in movies.Genre.cat.categories:
    list1.append(movies[movies.Genre == gen].BudgetMillion)
    mylabel.append(gen)
    
fig, ax=plt.subplots()    
fig.set_size_inches(11.7,8.27) # size of A4 paper
plt.hist(list1, bins=30, stacked=True, rwidth=1, label = mylabel)
plt.title('Movie Budget Distribution', fontsize=35, color='DarkBlue',\
         fontname='Console')
plt.ylabel('# of movies', fontsize=20, color='Red')
plt.yticks(fontsize=15)
plt.xlabel('Budget', fontsize=20, color='Green')
plt.xticks(fontsize=15)
plt.legend(frameon=True,fancybox=True,shadow=True,framealpha=1,prop={'size':15})
plt.show()
```
